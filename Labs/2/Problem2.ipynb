```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df2 = pd.read_csv("DF2", index_col = 0)
df2.head(10)
```

```python
df2.plot.scatter(x = '0', y = '1',s = 1)
```

The intuitive reasonsing for why (-1,1) is more outlying is because a linear regression model would have deviate from (-1,1) more than (5.5, 5)

```python
df2.hist(bins = 100)
```

We see that this distribution is clearly a multivariate guassian

```python
e_val, e_vec = np.linalg.eig(df2.cov())
```

```python
print(e_vec)
```

```python
print(e_val)
```

We can see that the first eigenvalue has much greater magnitue, thus directs the data much more.


```python
projected = df2@e_vec
```

```python
plt.scatter(projected[0], projected[1])
```

One can see that the first point now clearly is more variant
