On input k and XXXX, the below function returns the top k names from year XXXX:

```python
def popular_names(k, year):
    # Open the appropriate file
    df = pd.read_csv("Names/yob" + str(year) + ".txt", header = None)
    # Determine the k most populare names
    pop_names = []
    for i in range(len(df)):
        pop_names.append([i, df[2][i]])
    pop_names.sort(key = lambda x: x[1])
    names = []
    for i in range(len(df) - 1, len(df) - k - 1, -1):
        names.append(df[0][pop_names[i][0]])
    return names
```

On input Name, this function returns the frequency for men and women of the given Name:

```python
def sex_name_counts(name):
    counts = {"Men" : {},  "Women" : {}}
    # For each year
    for i in range(1880, 2015 + 1):   
    # Open the corresponding file
        df = pd.read_csv("Names/yob" + str(i) + ".txt", header = None)
        # Find the columns with name "name"
        df = df[df[0] == name]
        for j in range(len(df)):
            if df.iloc[j,1]== 'M':
                counts["Men"][str(i)]  = df.iloc[j,2]
            else:
                counts["Women"][str(i)]  = df.iloc[j,2]
    return counts
```

Modified above function to return relative frequency:

```python
# Calculate the total names for each year
totalNames = [0 for i in range(2015 + 1 - 1880)]
for i in range(0, (2015 + 1 - 1880)):
    # Open the file
    df = pd.read_csv("Names/yob" + str(1880 + i) + ".txt", header = None)
    total = 0
    for j in range(len(df)):
        total += df.iloc[i,2]
    totalNames[i] = total
```

```python
# Write the relative frequency method
def rel_freq(name):
    counts = {"Men" : {},  "Women" : {}}
    # Iterate through each year
    for i in range(1880, 2015 + 1):
        # Read the file
        df = pd.read_csv("Names/yob" + str(i) + ".txt", header = None)
        # Isolate by name
        df = df[df[0] == name]
        for j in range(len(df)):
            if df.iloc[j, 1] == "M":
                counts["Men"][str(i)] = df.iloc[j,2] / totalNames[i - 1880]
            else:
                counts["Women"][str(i)] = df.iloc[j,2] / totalNames[i - 1880]
    return counts
```

```python
john_rel_freq = rel_freq(name = "John")
```

Function to find names that used to be more popular for one gender, but became more popular for the other gender overtime:
Our algorithm loops through all years, 1880 to 2015, storing each name's initial "popular-sex" and looks for a "Switch" of popularity at any succeeding year.

```python
name_switches = {} # name : sex
switched_names = set()
# For each year
for i in range(1880, 2015 + 1):
    print(i)
    # Read the file
    df = pd.read_csv("Names/yob" + str(i) + ".txt", header = None)
    # Get the set of all the names from the dataframe
    names = set(df[0].unique())
    for name in names:
        rows = df.loc[df[0] == name]
        # If there is an entry for both sexes
        dominant_sex = 'M'
        if len(rows) == 2:
            # Determine the greater sex
            dominant_sex  = rows.iloc[0, 1] if (rows.iloc[0,2] > rows.iloc[1,2]) else rows.iloc[1, 1]
        else:
            dominant_sex = rows.iloc[0, 1]
        # If we have seen it before, check the last dominate, and note if it is different
        if name in name_switches and dominant_sex != name_switches[name]:
            switched_names.add(name)
        # If we have not seen it before note the sex
        else:
            name_switches[name] = dominant_sex
switched_names
```
